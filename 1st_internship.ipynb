IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

LOADING DATASET

```python
df1 = pd.read_csv("student-por.csv",sep = ';')
```

```python
df1.head()
```

```python
df2 = pd.read_csv("student-mat.csv",sep = ';')
```

```python
df2.head()
```

MERGING TWO DATASETS

```python
df = pd.concat([df1,df2],ignore_index = True)
```

VISUALIZATION


HIST OF GRADES

```python
df[['G1','G2','G3']].hist(bins = 40,figsize = (10,5))
```

STUDYTIME VS GRADES

```python
plt.title('G1')   
 
plt.xlabel('studytime')   
plt.ylabel('Grades')  
 
plt.bar(df.studytime, df.G1) 
plt.show()
```

```python
plt.title('G2')   
 
plt.xlabel('studytime')   
plt.ylabel('Grades')  
 
plt.bar(df.studytime, df.G2) 
plt.show()
```

```python
plt.title('G3')   
 
plt.xlabel('studytime')   
plt.ylabel('Grades')  
 
plt.bar(df.studytime, df.G3) 
plt.show()
```

SCATTER PLOT

```python
fig,axes = plt.subplots(1,3,figsize = (12,4))

axes[0].scatter(df['studytime'],df['G1'])
axes[0].set_title("studytime vs G1")
axes[0].set_xlabel("studytime")
axes[0].set_ylabel("G1")

axes[1].scatter(df['studytime'],df['G2'])
axes[1].set_title("studytime vs G2")
axes[1].set_xlabel("studytime")
axes[1].set_ylabel("G2")

axes[2].scatter(df['studytime'],df['G3'])
axes[2].set_title("studytime vs G3")
axes[2].set_xlabel("studytime")
axes[2].set_ylabel("G3")

plt.tight_layout()
plt.show()
```

MALE VS FEMALE AVERAGE SCORE

```python
avg_score = df.groupby('sex')['G3'].mean()
```

```python
avg_score.plot(kind = 'bar' )

plt.title("Average G3 Score by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Score")

plt.show()
```
